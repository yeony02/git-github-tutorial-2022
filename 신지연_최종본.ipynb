# 1-1. 2018~2021 국내/국제선 증감률 표



출처: 『항공통계』,국토교통부

코로나 확산 이전의 2년과 코로나 확산 이후 2년을 비교한 표이다. 코로나 확산이 시작된 이후 타국의 입국 금지 조치 및 국경 폐쇄 등 엄격한 여행 금지 조치로 인해 국제선의 경우 약 85퍼센트까지 감소된 모습을 볼 수 있다. 국내선의 경우 2020년 주춤하는 모습을 보이나 외국 여행 대신 국내 여행을 택하는 국민들로 인해 2021년이 되어 금세 회복되는 모습을 볼 수 있다.




```python
import pandas as pd
df1=pd.read_csv('/content/국내:국제선 여객 증감표.csv',index_col=0)
df1
```

#1-2. 2018~2021 국내/국제선 여객수 증감률 시각화 그래프



```python
fig = plt.figure()
ax = df1.plot(kind='bar',color=['cornflowerblue','plum'], figsize=(10, 8), legend=True,fontsize=13)
ax.set_title('<국내/국제선 여객수 증감률>', fontsize=30)
ax.set_xlabel('년도', fontsize=25)
ax.set_ylabel('증감률(%)', fontsize=25)
ax.legend(['국제선','국내선'], fontsize=15)
plt.xticks(rotation = 0, fontsize = '17')
plt.yticks(rotation = 0, fontsize = '17')
plt.axhline(0,0,5,color='gray',linestyle='--')
%config InlineBackend.figure_format='retina'
```

#2-1. 국내 8개 항공사별 운항(편)수

출처:『항공통계』, 한국항공공사, 인천국제공항공사

국내 8개의 항공사별 4월의 여객기 운항 편수를 알 수 있는 데이터이다. 국내에 코로나가 발생하기 이전 시점인 2019년 이전과 2020년 이후의 운항 편수가 대형 항공사일수록 확연한 차이를 보인다. 

```python
import pandas as pd
df2=pd.read_excel('/content/항공사별 운항, 여객 통계(2018~2021).xlsx',index_col=0)

df2
```

# 2-2. 국내 8개 항공사별 운항(편)수 시각화 그래프

```python
df2=df2.transpose()
```

```python
fig = plt.figure()
ax = df2.plot(kind='line',color=['skyblue','red','darkgray','moccasin','navy','yellowgreen','orchid','firebrick','gray'], figsize=(10, 8), legend=True,fontsize=13)
ax.set_title('<항공사별 운항(편)수 >', fontsize=28)
plt.xticks([2018.04, 2019.04, 2020.04, 2021.04])
ax.set_xlabel('연도', fontsize=25)
ax.set_ylabel('운항 수', fontsize=25)
ax.legend(['아시아나','에어부산','에어인천','에어서울','제주항공','진에어','대한항공','티웨이항공'], fontsize=11)
plt.xticks(fontsize = '16')
plt.vlines(2019.8, 1000, 17000, color='firebrick', linestyle='--', linewidth=3)
plt.text(2019.5, 17500, '국내 코로나19 발생',color='firebrick',fontsize=15)

%config InlineBackend.figure_format='retina'
```
